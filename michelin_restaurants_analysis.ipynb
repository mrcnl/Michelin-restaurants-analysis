# Michelin restaurants in the world

The subject of this notebook is the analysis of restaurants that can pride themselves on a star (or stars) awarded by Michelin. The dataset is downloaded from https://www.kaggle.com/jackywang529/michelin-restaurants and it is divided into three csv files. Each of them shows the restaurants (and their details) which were awarded with one, two and three stars respectively.

It is necessary to announce that the dataset does not contain the list of all of the restaurants with Michelin star. In the list there are no restauarants that are located i.a. in Germany, Spain or Italy.

## Data cleaning

```python
import pandas as pd
```

```python
michelin_1_df = pd.read_csv('Csv_files/michelin1.csv')
michelin_2_df = pd.read_csv('Csv_files/michelin2.csv')
michelin_3_df = pd.read_csv('Csv_files/michelin3.csv')
```

```python
michelin_1_df
```

```python
michelin_2_df
```

```python
michelin_3_df
```

```python
michelin_1_df['number_of_stars'] = 1
michelin_2_df['number_of_stars'] = 2
michelin_3_df['number_of_stars'] = 3

michelin_df = pd.concat([michelin_1_df, michelin_2_df,
                         michelin_3_df], ignore_index = True)

del michelin_df['url']
del michelin_df['year']
del michelin_df['zipCode']

michelin_df
```

**Adding 'continent' column**

```python
import numpy as np
conditions = [(michelin_df['latitude'] >= 1.27) & (michelin_df['latitude'] <= 77.72) & (michelin_df['longitude'] >= 26.08) & (michelin_df['longitude'] <= 169.67),
              (michelin_df['latitude'] >= 34.83) & (michelin_df['latitude'] <= 81.9) & (michelin_df['longitude'] >= -39.47) & (michelin_df['longitude'] <= 69.08),
              (michelin_df['latitude'] >= -59.48) & (michelin_df['latitude'] <= 13.4) & (michelin_df['longitude'] >= -109.45) & (michelin_df['longitude'] <= -26.27),
              (michelin_df['latitude'] >= 5.53) & (michelin_df['latitude'] <= 83.67) & (michelin_df['longitude'] >= -175.47) & (michelin_df['longitude'] <= -11.32),
              (michelin_df['latitude'] >= -39) & (michelin_df['latitude'] <= -11) & (michelin_df['longitude'] >= 113) & (michelin_df['longitude'] <= 153)]

values = ['Asia', 'Europe', 'South America', 'North America', 'Australia']

michelin_df.insert(5, 'continent', np.select(conditions, values, default = 'Somewhere in the ocean'))

michelin_df
```

```python
michelin_df['continent'].unique()
```

**Organizing the 'region' column and changing it into the 'country' column**

```python
michelin_df['region'].unique()
```

```python
michelin_df['region'].replace(['New York City','Chicago', 'California', 'Washington DC', 'Rio de Janeiro', 'Sao Paulo', 'Macau', 'Hong Kong', 'Taipei'],
                              ['United States', 'United States', 'United States', 'United States', 'Brazil', 'Brazil', 'China', 'China', 'Taiwan'],
                              inplace = True)

michelin_df.rename(columns = {'region': 'country'}, inplace = True)

michelin_df['country'].unique()
```

**Organizing the 'city' column**

```python
michelin_df['city'].unique()
```

```python
michelin_df['city'].replace(['Rio de Janeiro.*', 'São Paulo.*', '.*Saint Helier.*', 'Blackrock', 'City Centre', 'Washington.*'],
                            ['Rio de Janeiro', 'Sao Paulo', 'Saint Helier', 'Dublin', 'Dublin', 'Washington'],
                            regex = True, inplace = True)

michelin_df['city'].replace(['.*London.*', 'Kew', 'Chiswick', 'Hammersmith', 'Kensington', 'Marylebone', "Regent's Park",
                             'Fulham', 'Bloomsbury', 'Mayfair', 'Soho', 'Belgravia', 'Westminster', 'Chelsea', "Saint James's",
                             'Victoria', 'Clerkenwell', 'London', 'Shoreditch', 'Finsbury', 'Spitalfields', 'Bermondsey',
                             'Clapham Common', 'Wandsworth', 'North Kensington', 'Hyde Park'  ],
                            'London',
                            regex = True, inplace = True)

michelin_df['city'].unique()
```

```python
michelin_df.info()
```

**Filling in the NaN values in the 'city' column**

```python
michelin_df[michelin_df['city'].isna()]
```

```python
michelin_df['city'].fillna('Hong Kong', inplace = True)
```

**Organizing the 'cuisine' column**

```python
michelin_df['cuisine'].unique()
```

```python
michelin_df['cuisine'].replace(['.*[Cc]hinese.*', '.*Cantonese', 'Dim Sum', 'Shanghainese', 'Noodles and congee',
                                'Taizhou', '.*Sichuan.*', 'Fujian', 'Hang Zhou', 'Chinese-huai yang'],
                               'Chinese',
                               inplace = True, regex = True)

michelin_df['cuisine'].unique()
```

```python
michelin_df.groupby('cuisine')[['cuisine']].count()
```

```python
michelin_df.shape
```

```python
for i in range(michelin_df.shape[0]):
    michelin_df.at[i, 'cuisine'] = michelin_df.at[i, 'cuisine'].capitalize()
    
michelin_df['cuisine'].replace(['.*[Ee]uropean.*', '.*[Ff]rench.*', '.*[Bb]ritish.*', '.*[Ii]talian.*', '.*[Jj]apanese.*',
                                'Teppanyaki', '.*[Mm]editerranean.*', '.*[Tt]hai.*',  '.*[Aa]sian.*', '.*[Kk]orean.*',
                                'Temple cuisine', '.*[Mm]odern.*'],
                               ['European', 'French', 'British', 'Italian', 'Japanese', 'Japanese', 'Mediterranean',
                                'Thai', 'Asian', 'Korean', 'Korean', 'Modern'],                                
                                regex = True, inplace = True)

michelin_df['cuisine'].unique()
```

```python
michelin_df.groupby('cuisine')[['cuisine']].count()
```

**Organizing the 'price' column**

```python
michelin_df['price'].unique()
```

```python
michelin_df['price'].fillna(0, inplace = True)

michelin_df['price'].replace(['$', '$$', '$$$', '$$$$', '$$$$$'],
                             [1, 2, 3, 4, 5], inplace = True)

michelin_df.rename(columns = {'price': 'price_from_1_to_5'}, inplace = True)

michelin_df['price_from_1_to_5'].unique()
```

## Exploratory Analysis and Visualization

```python
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style('darkgrid')
matplotlib.rcParams['font.size'] = 16
matplotlib.rcParams['figure.facecolor'] = '#00000000'
```

### Countries with the greatest number of restaurants with Michelin star(s)

```python
michelin_countries_df = michelin_df.groupby('country')[['country']].count()

michelin_countries_df = michelin_countries_df[michelin_countries_df['country'] > 10]

michelin_countries_df.rename(columns = {'country': 'number_of_restaurants'}, inplace = True),

michelin_countries_df
```

**In order to tailor that DataFrame to other DataFrame there is a need to multiply the number of rows**

```python
michelin_countries_df1 = pd.DataFrame(np.repeat(michelin_countries_df.values,3,axis=0))
michelin_countries_df1.columns = michelin_countries_df.columns

michelin_countries_df = michelin_countries_df1.copy()
del michelin_countries_df1

michelin_countries_df
```

**In order to select only the countries with at least 10 restaurants the 'number_of_restaurants_in_country' column is necessary.**

```python
michelin_df_copy = michelin_df.copy()
michelin_df_copy['number_of_restaurants_in_country'] = michelin_df_copy.groupby('country')['country'].transform('count')
michelin_df_copy = michelin_df_copy[michelin_df_copy['number_of_restaurants_in_country'] >= 10].reset_index()
michelin_df_copy
```

```python
michelin_stars_df = michelin_df_copy.groupby(['country', 'number_of_stars'])[['country']].count()

michelin_stars_df.rename(columns = {'country': 'number_of_n_star_restaurants'},
                           index = {'United States': 'US', 'United Kingdom': 'UK' }, inplace = True)

michelin_stars_df
```

```python
levels = [michelin_stars_df.index.levels[0].values, michelin_stars_df.index.levels[1].values]
new_index = pd.MultiIndex.from_product(levels, names=michelin_stars_df.index.names)
michelin_stars_df = michelin_stars_df.reindex(new_index, fill_value=0)
michelin_stars_df
```

```python
michelin_stars_df = michelin_stars_df.reset_index(level=[0,1])
michelin_stars_df
```

**Now the DataFrame that specifies the number of restaurants with each number of stars in each country is concatenated with the DataFrame that specifies the number of all of the restaurants with Michelin stars in each country**

```python
michelin_countries_stars_df = pd.concat([michelin_stars_df, michelin_countries_df], axis = 1)
michelin_countries_stars_df.sort_values(['number_of_restaurants', 'country', 'number_of_stars'],
                                        ascending = [False, True, True],
                                        inplace = True,)

michelin_countries_stars_df
```

```python
fig, ax = plt.subplots(figsize = (20,12))
ax.set_title('Number of Michelin 1-, 2- and 3-stars restaurants in the world', fontsize = 25)
ax.bar(michelin_countries_stars_df['country'].unique(), michelin_countries_stars_df['number_of_n_star_restaurants'].values[::3])
ax.bar(michelin_countries_stars_df['country'].unique(), michelin_countries_stars_df['number_of_n_star_restaurants'].values[1::3], bottom = michelin_countries_stars_df['number_of_n_star_restaurants'].values[::3])
ax.bar(michelin_countries_stars_df['country'].unique(), michelin_countries_stars_df['number_of_n_star_restaurants'].values[2::3], bottom = michelin_countries_stars_df['number_of_n_star_restaurants'].values[::3] + michelin_countries_stars_df['number_of_n_star_restaurants'].values[1::3])
ax.legend(['1 star', '2 stars', '3 stars'])

plt.show()
```

### Geographic distribution of the restaurants in every continent

```python
michelin_n_america_df = michelin_df[michelin_df['continent'] == 'North America'][['latitude', 'longitude']].reset_index()
michelin_s_america_df = michelin_df[michelin_df['continent'] == 'South America'][['latitude', 'longitude']].reset_index()
michelin_europe_df = michelin_df[michelin_df['continent'] == 'Europe'][['latitude', 'longitude']].reset_index()
michelin_asia_df = michelin_df[michelin_df['continent'] == 'Asia'][['latitude', 'longitude']].reset_index()
```

**A good way to prepare pivot tables from above DataFrames is to aggregate data into specific bins**

```python
bins_n_america1 = np.arange(26,50)
bins_n_america2 = np.arange(-125,-65)
michelin_n_america_df['latitude_bins'] = pd.cut(michelin_n_america_df['latitude'], bins_n_america1)
michelin_n_america_df['longitude_bins'] = pd.cut(michelin_n_america_df['longitude'], bins_n_america2)

bins_s_america1 = np.arange(-42,6)
bins_s_america2 = np.arange(-72,-34)
michelin_s_america_df['latitude_bins'] = pd.cut(michelin_s_america_df['latitude'], bins_s_america1)
michelin_s_america_df['longitude_bins'] = pd.cut(michelin_s_america_df['longitude'], bins_s_america2)

bins_europe1 = np.arange(34,70)
bins_europe2 = np.arange(-24,42)
michelin_europe_df['latitude_bins'] = pd.cut(michelin_europe_df['latitude'], bins_europe1)
michelin_europe_df['longitude_bins'] = pd.cut(michelin_europe_df['longitude'], bins_europe2)

bins_asia1 = np.arange(42)
bins_asia2 = np.arange(90,135)
michelin_asia_df['latitude_bins'] = pd.cut(michelin_asia_df['latitude'], bins_asia1)
michelin_asia_df['longitude_bins'] = pd.cut(michelin_asia_df['longitude'], bins_asia2)
```

```python
michelin_n_america_df = michelin_n_america_df.groupby(['latitude_bins', 'longitude_bins'])[['latitude_bins']].count()

michelin_n_america_df.rename(columns = {'latitude_bins': 'number_of_restaurants'}, inplace = True)

michelin_n_america_df = michelin_n_america_df.reset_index(level=[0,1])

michelin_n_america_df['number_of_restaurants'].fillna(0, inplace = True)

michelin_n_america_df
```

```python
michelin_s_america_df = michelin_s_america_df.groupby(['latitude_bins', 'longitude_bins'])[['latitude_bins']].count()

michelin_s_america_df.rename(columns = {'latitude_bins': 'number_of_restaurants'}, inplace = True)

michelin_s_america_df = michelin_s_america_df.reset_index(level=[0,1])

michelin_s_america_df['number_of_restaurants'].fillna(0, inplace = True)

michelin_s_america_df
```

```python
michelin_europe_df = michelin_europe_df.groupby(['latitude_bins', 'longitude_bins'])[['latitude_bins']].count()

michelin_europe_df.rename(columns = {'latitude_bins': 'number_of_restaurants'}, inplace = True)

michelin_europe_df = michelin_europe_df.reset_index(level=[0,1])

michelin_europe_df['number_of_restaurants'].fillna(0, inplace = True)

michelin_europe_df
```

```python
michelin_asia_df = michelin_asia_df.groupby(['latitude_bins', 'longitude_bins'])[['latitude_bins']].count()

michelin_asia_df.rename(columns = {'latitude_bins': 'number_of_restaurants'}, inplace = True)

michelin_asia_df = michelin_asia_df.reset_index(level=[0,1])

michelin_asia_df['number_of_restaurants'].fillna(0, inplace = True)

michelin_asia_df
```

### Q3: The average price of various kinds of cuisine

```python
michelin_prices_df.info()
```

```python
michelin_cuisine_price = michelin_prices_df.groupby('cuisine').agg({'cuisine': 'count', 'price_from_1_to_5': 'mean'})

michelin_cuisine_price = michelin_cuisine_price[michelin_cuisine_price['cuisine'] >= 10]

michelin_cuisine_price.rename(columns = {'cuisine': 'number_of_restaurants'}, inplace = True)

michelin_cuisine_price['price_from_1_to_5'] = round(michelin_cuisine_price['price_from_1_to_5'],1)

michelin_cuisine_price.sort_values('price_from_1_to_5', ascending = False, inplace = True)

michelin_cuisine_price
```

**Among kinds of cuisine that are characteristic for at least 10 restaurants creative cuisine is the most expensive one and Chinese cuisine is the cheapest one.**

### Q4: The average price of restaurants in particular cities, countries and continents

```python
michelin_city_price = michelin_prices_df.groupby('city').agg({'city': 'count', 'price_from_1_to_5': 'mean'})

michelin_city_price = michelin_city_price[michelin_city_price['city'] >= 10]

michelin_city_price.rename(columns = {'city': 'number_of_restaurants'}, inplace = True)

michelin_city_price['price_from_1_to_5'] = round(michelin_city_price['price_from_1_to_5'],1)

michelin_city_price.sort_values('price_from_1_to_5', ascending = False, inplace = True)

michelin_city_price
```

**The city with the highest average price of restaurants with Michelin star(s) is Wien. The city with the lowest average price is Singapore.**

```python
michelin_country_price = michelin_prices_df.groupby('country').agg({'country': 'count', 'price_from_1_to_5': 'mean'})

michelin_country_price = michelin_country_price[michelin_country_price['country'] >= 10]

michelin_country_price.rename(columns = {'country': 'number_of_restaurants'}, inplace = True)

michelin_country_price['price_from_1_to_5'] = round(michelin_country_price['price_from_1_to_5'],1)

michelin_country_price.sort_values('price_from_1_to_5', ascending = False, inplace = True)

michelin_country_price
```

**The country with the highest average price of restaurants with Michelin star(s) is Austria. The country with the lowest average price is Singapore.**

```python
michelin_continent_price = michelin_prices_df.groupby('continent').agg({'continent': 'count', 'price_from_1_to_5': 'mean'})

michelin_continent_price.rename(columns = {'continent': 'number_of_restaurants'}, inplace = True)

michelin_continent_price['price_from_1_to_5'] = round(michelin_continent_price['price_from_1_to_5'],1)

michelin_continent_price.sort_values('price_from_1_to_5', ascending = False, inplace = True)

michelin_continent_price
```

**The continent with the highest average price of restaurants with Michelin star(s) is South America. The continent with the lowest average price is Asia.**

**However, in South America  there are only 18 restaurants with Michelin star which is much less than in remaining continents.**
